Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
```

```python
X = pd.DataFrame(boston["data"], columns=boston["feature_names"])
```

```python
y = pd.DataFrame(boston["target"], columns=["price"])
```

```python
#X, y = load_boston(return_X_y=True)
print('X' , X.shape)
print('y', y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print('X_train', X_train.shape)
print('X_test', X_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
      
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
lr.coef_.shape
```

```python
y_predict = lr.predict(X_test)
y_predict.shape
```

```python
r2_lr = r2_score(y_test, y_predict)
r2_lr
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_csv('./creditcard.csv',sep=',')
df.tail(5)
```

```python
df['Class'].value_counts(dropna=False,normalize=True)
```

```python
df.info()
```

```python
df.head(10)
```

```python
y = df[['Class']].iloc[:,0]
type(y)
```

```python
X = df.iloc[:, 0:30]
X.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
print('X_train', X_train.shape)
print('X_test', X_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

```python
parameters = [{'n_estimators': [10, 15], 'max_features': np.arange(3, 5), 'max_depth': np.arange(4, 7)}]
```

```python
GSCV = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
GSCV.fit(X_train,y_train)
```

```python
GSCV.best_params_
```

```python
y_pred_proba = GSCV.predict_proba(X_test)[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba, average=None)
```
